```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("D:\python\img1.jpg")
plt.imshow(img1[:,:,::-1])
plt.show
```

```python
analyze = DeepFace.analyze(img1,actions = ['emotion'])
print(analyze)
```

```python
try: 
    dominant_emotion = analyze[1]['dominant_emotion']
    # as the second analysis is more accurate than the first
except IndexError:
    dominant_emotion = analyze[0]['dominant_emotion']
    # if only 1 result is generated
    
print(dominant_emotion)
```
